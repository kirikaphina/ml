```python
import pandas as pd
import numpy as np
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/msoma-org/endangered-languages/main/data.csv'
df = pd.read_csv(url,index_col=0)
df
```

```python
df.head()
```

```python
df.tail()
```

```python
plt.figure(figsize=(16,6))
sns.lineplot(data=df)
```

```python
missing_values_count = df.isnull().sum()
missing_values_count[0:10]
```

```python
total_cells = np.product(df.shape)
total_missing = missing_values_count.sum()

percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

```python
#The number of speakers will be on the y-Axis, and the degree of endangerment on the x-Axis.
plt.figure(figsize=(10,6))

plt.title("the number of languages by degree of endangerment")

sns.barplot(x=df['Degree of endangerment'], y=df['Number of speakers'])

plt.ylabel("The number of speakers")
```
